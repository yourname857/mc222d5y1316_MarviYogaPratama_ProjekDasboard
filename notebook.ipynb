## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh suhu terhadap jumlah penyewaan sepeda?

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=day_df['temp'], y=day_df['cnt'], alpha=0.5)
plt.title('Hubungan antara Suhu dan Jumlah Penyewaan Sepeda')
plt.xlabel('Suhu')
plt.ylabel('Jumlah Penyewaan')
plt.show()
```

**Insight:**
- Membuat scatter plot antara suhu (temp) dan jumlah penyewaan sepeda (cnt)

yang dimana hasil visualisasi menunjukkan bahwa suhu memiliki korelasi positif terhadap jumlah penyewaan sepeda, dan ketika suhu meningkat, jumlah penyewaan juga cenderung meningkat.

### Pertanyaan 2:Bagaimana tren penyewaan sepeda sepanjang hari?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=hour_df['hr'], y=hour_df['cnt'], ci=None)
plt.title('Tren Penyewaan Sepeda sepanjang Hari')
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan')
plt.show()
```

**Insight:**
Mengapa grafik ini penting?
Karena untuk memahami pola penyewaan sepanjang hari. Bisa melihat kapan puncak penyewaan terjadi.
- Grafik menunjukkan bahwa jumlah penyewaan sepeda meningkat pada jam-jam tertentu. Biasanya puncak terjadi di pagi hari ( jam berangkat kerja ) dan sore hari (jam pulang kerja).

## Analisis Lanjutan (Opsional)

```python
import datetime as dt

hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

latest_date = hour_df['dteday'].max()

rfm_df = hour_df.groupby(['dteday']).agg({
    'cnt': 'sum'
}).reset_index()
rfm_df['Recency'] = (latest_date - rfm_df['dteday']).dt.days
rfm_df['Frequency'] = hour_df.groupby('dteday')['cnt'].count().values
rfm_df['Monetary'] = hour_df.groupby('dteday')['cnt'].sum().values
print(rfm_df.describe())
plt.figure(figsize=(12, 6))
sns.histplot(rfm_df['Recency'], bins=30, kde=True)
plt.title('Distribusi Recency')
plt.show()
plt.figure(figsize=(12, 6))
sns.histplot(rfm_df['Frequency'], bins=30, kde=True)
plt.title('Distribusi Frequency')
plt.show()
plt.figure(figsize=(12, 6))
sns.histplot(rfm_df['Monetary'], bins=30, kde=True)
plt.title('Distribusi Monetary')
plt.show()
```

## Hasil Analisis RFM yang Saya Lakukan
Dalam analisis RFM yang saya kerjakan, saya menggunakan tiga parameter utama:


1.   Recency (R): Saya mengukur seberapa baru aktivitas pelanggan dengan menghitung selisih hari antara tanggal terakhir dalam dataset dan tanggal transaksi tiap pelanggan. Ini membantu saya memahami pola kunjungan terbaru pelanggan.
2.   Frequency (F): Untuk aspek ini, saya menghitung berapa kali pelanggan menyewa sepeda dengan cara menjumlah semua transaksi harian. Metode ini memungkinkan saya melihat loyalitas pelanggan dari sisi keseringan penyewaan.
3. Monetary (M): Dalam parameter ini, saya menjumlahkan total penyewaan sepeda per hari untuk mendapatkan gambaran volume bisnis secara keseluruhan.

Dari analisis yang saya lakukan, saya menemukan beberapa pola menarik. Distribusi Recency memberikan saya pemahaman tentang waktu terakhir penyewaan terjadi. Sementara itu, distribusi Frequency menunjukkan kepada saya pola keseringan penyewaan, dan distribusi Monetary memperlihatkan total volume penyewaan harian.
Dengan metode ini, saya dapat mengidentifikasi segmen pelanggan dan trend penyewaan sepeda yang bisa menjadi dasar pengambilan keputusan bisnis selanjutnya.
